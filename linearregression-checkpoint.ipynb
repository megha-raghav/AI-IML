```python
# Importing important libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

```

```python
#importing the data set
data=pd.read_csv("C:\dataset\Car_Purchasing_Data.csv",encoding = "ISO-8859-1")
data.head()
```

```python
# X= Customer Name, Customer e-mail, Country, Gender, Age, Annual Salary, Credit Card Debt, Net Worth 
# y= Car Purchase Amount
# Now dividing the dataset into trainnig data and testing data(for this we have to import train_test_split package from sklearn.model_selection library)
X=data[['Gender', 'Age', 'Annual Salary', 'Credit Card Debt', 'Net Worth']]
Y=data[['Car Purchase Amount']]
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
X_train
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# creating an object of class LinearRegression( as we import before from library sklear.linear_model)
lrmodel=LinearRegression()
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
# Now train a model
lrmodel.fit(X_train,Y_train)
```

```python
lrmodel.predict(X_test)

```

```python
# to find the accuracy
lrmodel.score(X_train,Y_train)
```
